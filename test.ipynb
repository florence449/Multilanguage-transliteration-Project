```python
# 함수 임포트하기
from ipa_to_kor import *
import pandas as pd
import random

def random_print(data):
    num = random.randint(0, len(data)-1)
    print(f"{num}번째 값")
    print(f"원어 표기 : \t{data.loc[num, 'name']}")
    print(f"ipa 표기 : \t{data.loc[num, 'ipa_name']}")
    print(f"한글 표기 : \t{data.loc[num, 'kor_name']}")
```

```python
data = pd.read_csv('test_data/fr_names_ipa.csv')
```

```python
data['kor_name'] = data['ipa_name'].apply(convert_to_kor, language= 3)
```

```python
random_print(data)
```

```python
data.to_csv("fr_test.csv", index=False)
```
